## 1.1- Business Problem - Introduction and Motivation

For reputational and financial reasons, hospitals tend to avoid readmissions. Readmission is a healthcare metric that tracks patients who are re-admitted to a hospital **within 30 days**. This is generally avoided by hospitals and is in the best interest of patients due to many complications, such as reduced coverage during readmissions from insurances like Medicare, which can cause financial issues for hospitals and patients themselves (2012). Hence, we believe that by analyzing the most prominent groups of readmitted patients, we can lower the readmission rates by identifying those that might require additional outpatient treatment, longer stays, or other preventative measures.

-----

**Project objective/proposal**:

Our project objective/proposal is analyzing demographic, diabetic diagnoses, and medication histories to understand their impact on readmission types.

##1.2 - Data Source - Description

**Download URL**: https://data.mendeley.com/datasets/nntck7ddgt/2

Through the data from Clinical and Translational Research at Virginia Commonwealth University, which is linked [HERE](https://data.mendeley.com/datasets/nntck7ddgt/2), we're able to explore 10 years (1999-2008) of clinical care data at 130 US hospitals and integrated delivery networks. This dataset includes information on diabetic inpatient encounters ranging from 1 to 14 days and whether the encountered patients are readmitted (No, > 30 days, < 30 days). There are 55 variables and 101,766 instances, including some with missing values marked as '?'. The variables can generally be divided into four portions:

- **Demographic/basic information:**
  - `encounter_id`: Unique identifier for the encounter
  - `patient_nbr`: Unique identifier of the patient
  - `race`: Patient's race (Mapped to numercal values from 0 to 6)
  - `gender`: Patient's gender (Male/Female)
  - `age`: Age grouped in 10-year intervals from 0 to 100
  - `weight`: Weight in pounds
      
-  **Diagnosis:**
  - `admission_type_id`: Integer identifier that corresponds to different types of admission (mapping in Appendix A.)
  - `discharge_disposition_id`: Integer identifier that corresponds to different types of discharge conditions (mapping in Appendix Section)
  - `admission_source_id`
  - `time_in_hospital`: Number of days between admission and discharge
  - `payer_code`: Unique identifier (key) that indicates who is paying, includes self-pay
  - `medical_specialty`: Specialty of the admitting doc
  - `num_lab_procedures`: Number of lab tests performed during stay
  - `num_procedures`: Number of prcedures (other than lab tests) during stay
  - `num_medications`: Number of distinct generic meds given during stay
  - `number_outpatient`: Number of outpatient(service/treatment that does not require hospitalization) visits the patient had in the year preceding the stay
  - `number_emergency`: Number of emergency visits of the patient preceding the stay
  - `number_inpatient`: Number of inpatient visits (required hospitalization) in the year preceding the stay
  - `diag_1, diag_2, diag_3`: the primary and two secondary diagnosis of each patient. Due to the high number of possible diagnosis, we first group them into higher level diseases groups according to [2008 ICD-9-CM Data](http://www.icd9data.com/2008/Volume1/default.htm)
  - `number_diagnoses`: Total number of diagnosis
  - `max_glu_serum`, A1Cresult: Tests for blood sugar the day of and in the past three months respectively.

- **Diabetic Medications: list of diabetic medicines**
    - `metformin, repaglinide, nateglinide, chlorpropamide, glimepiride, acetohexamide, glipizide, glyburide, tolbutamide, pioglitazone, rosiglitazone, acarbose, miglitol, troglitazone, tolazamide, examide, citoglipton, insulin, glyburide-metformin, glipizide-metformin, glimepiride-pioglitazone, metformin-rosiglitazone, metformin-pioglitazon`

    - `change`: Indicates if there is a change in diabetic medication (either dosage or generic name) (Ch/No)
    - `diabetesMed`: Indicates if there was any diabetic medications prescribed (yes/no)
        
- **Readmission:**
    - `readmitted`: if the patient will be readmitted into the same hospital (<30/ >30/ No)

```python
#Group diagnosis into higher level groups
def diag_to_cat(value):
    if 'V' in value:
        return 'Supplementary_Classification_Of_Factors_Influencing_Health_Status_And_Contact_With_Health_Services'
    elif 'E' in value:
        return 'Supplementary_Classification_Of_External_Causes_Of_Injury_And_Poisoning'
    elif '?' in value:
        return None
    value = float(value)
    value = int(value)
    if value in range(1, 140):
        return 'Infectious_And_Parasitic_Diseases'
    elif value in range(140, 240):
        return 'Neoplasms'
    elif value in range(240, 280):
        return 'Endocrine_Nutritional_And_Metabolic_Diseases_And_Immunity_Disorders'
    elif value in range(280, 290):
        return 'Diseases_Of_The_Blood_And_Blood-Forming_Organs'
    elif value in range(290, 320):
        return 'Mental_Disorders'
    elif value in range(320, 390):
        return 'Diseases_Of_The_Nervous_System_And_Sense_Organs'
    elif value in range(390, 460):
        return 'Diseases_Of_The_Circulatory_System'
    elif value in range(460, 520):
        return 'Diseases_Of_The_Respiratory_System'
    elif value in range(520, 580):
        return 'Diseases_Of_The_Digestive_System'
    elif value in range(580, 630):
        return 'Diseases_Of_The_Genitourinary_System'
    elif value in range(630, 680):
        return 'Complications_Of_Pregnancy_Childbirth_And_The_Puerperium'
    elif value in range(680, 710):
        return 'Diseases_Of_The_Skin_And_Subcutaneous_Tissue'
    elif value in range(710, 740):
        return 'Diseases_Of_The_Musculoskeletal_System_And_Connective_Tissue'
    elif value in range(740, 760):
        return 'Congential_Anomalies'
    elif value in range(760, 780):
        return 'Certain_Conditions_Originating_In_The_Perinatal Period'
    elif value in range(780, 800):
        return 'Symptoms_Signs_And Ill-Defined_Conditions'
    elif value in range(800, 1000):
        return 'Injury_And_Poisoning'
    else:
        return 'Not in Dictionary'
```

```python
#Create new columns for the higher level disease groups:
df_clean['diag_1_cat'] = df_clean['diag_1'].apply(diag_to_cat)
df_clean['diag_2_cat'] = df_clean['diag_2'].apply(diag_to_cat)
df_clean['diag_3_cat'] = df_clean['diag_3'].apply(diag_to_cat)

#Drop old columns:
df_clean = df_clean.drop(columns= ['diag_1', 'diag_2', 'diag_3'])
```

```python
#Mapping Prescriptions to yes(1) or no(0) prescription:
prescription_mapping = {'Down':1, 'Up':1, 'Steady':1, 'No':0}

#Create list of columns with the prescription to be transformed:
prescription_list = list(df.loc[:,'metformin': 'metformin-pioglitazone'].columns)

#Apply map:
for prescription in prescription_list:
  df_clean[prescription] = df_clean[prescription].replace(prescription_mapping)
```

```python
#Start a list of lists to hold the medication name and its % use
med_avg = [[],[]]
for i in df_clean[prescription_list].columns: #Calculate % use and append to list
  med_avg[0].append(i)
  med_avg[1].append((df_clean[i].sum()/len(df_clean[i])))
```

```python
#Drop columns with less than 0.1% use:
med_threshold = 0.001
med_col_drop = []

for n in range(len(prescription_list)):
  if med_avg[1][n] < med_threshold:
    med_col_drop.append(med_avg[0][n])
df_clean.drop(labels=med_col_drop, axis=1, inplace=True)
```

**Key observations for Graph 1 - A1Cresult by Age and Gender:**


---


1.   The graph illustrates that A1C abnormal results decrease with age, which may seem counterintuitive. Older patients tend to have better A1C readings, possibly because they adhere to their medication regimen more effectively, resulting in fewer abnormal A1C results.

#5 - Data Preparation for ML


---

After conducting our exploratory analysis, we identified a few variables with missing values that we will address in this section. Additionally, some variables contain non-numeric values and had to be transformed; we have chosen to one-hot encode them to better prepare the data for our predictive models.

```python
#Create a copy of the df used for data exploration
new_df = df_clean.copy()

#columns to fix: discharge_disposition_id, admission_source_id,diag_1_cat, diag_2_cat, diag_3_cat, race, gender, max_glu_serum, A1Cresult
#new_df.isna().sum() (commented out to save space)
```

<a name="5.1"></a>
##5.1 - Filling nulls, one-hot encoding for *race and gender*

* The 'gender' column was renamed to 'female', where female=`1`, male=`0`

```python
#Renaming gender column for clarity
new_df.rename(columns={'gender':'female'}, inplace=True)

#mode race and gender, which are missing 2239 and 3 values
for i in ['race', 'female']:
  new_df[i].fillna(new_df[i].mode()[0], inplace= True)

new_df=pd.get_dummies(new_df, columns=['race'])
```

<a name="5.2"></a>
## 5.2 - Dropping *'max_glu_serum', 'A1Cresult'*

* Due to a high number of missing values we decided to drop `'max_glu_serum'` and `'A1Cresult'`

```python
#Dropping max_glu_serum and A1Cresult due to large missing values
new_df = new_df.drop(['max_glu_serum', 'A1Cresult'], axis=1)
```

<a name="5.3"></a>
## 5.3 - Filling nulls & one-hot-encoding for *'discharge_disposition_id', 'admission_type_id', 'admission_source_id'*

```python
#Select columns to fill
columns_to_fill= ['discharge_disposition_id', 'admission_type_id', 'admission_source_id']

for i in columns_to_fill:
  new_df[i].fillna(new_df[i].mode()[0], inplace=True)

# To remove the numerical bias, we'll do one hot encodings for these mapped columns, see Appendix for mapping definition
for i in columns_to_fill:
    dummies = pd.get_dummies(new_df[i], prefix=i)
    new_df = pd.concat([new_df, dummies], axis=1)

new_df.drop(columns=columns_to_fill, inplace=True)
```

<a name="5.4"></a>
##5.4 - One-hot encoding for 'age'

* The `'age'` variable was transformed into binary columns representing age ranges, such as 0-10 and 10-20

```python
new_df= pd.get_dummies(new_df, columns=['age'])
```

<a name="5.5"></a>
##5.5 - One-hot encoding for 'diag_1_cat', 'diag_2_cat', 'diag_3_cat'

- We found out that diag 1, 2, and 3 are not time-series data, rather, they represent simultaneous diagnoses. Hence, we will be combining them through the 18 unique diagnosis categories

```python
#preparing them for merging as dummy columns
new_df.rename(columns={'diag_1_cat': 'diag_1_cat', 'diag_2_cat': 'diag_1_cat', 'diag_3_cat': 'diag_1_cat'}, inplace=True)
new_df1= pd.DataFrame(pd.get_dummies(new_df['diag_1_cat']))
```

```python
new_df1.columns = new_df1.columns.str.replace('diag_1_cat_', '')
# Sum columns with the same names
summed_df = new_df1.groupby(new_df1.columns, axis=1).sum()
```

```python
# Display the updated DataFrame
final_df=pd.concat([new_df, summed_df], axis=1)
final_df.drop(['diag_1_cat'],  axis=1, inplace= True)
```

<a name="5.6"></a>
## 5.6 - Final dataset (`final_df`)
- No null values
- All columns in numerical values
- 100,114 Observations and 71 columns (including all one-hot encoded values)

```python
final_df.drop('patient_nbr', axis=1, inplace= True)

for i in final_df.columns:
  final_df[i].astype(int)

final_df = final_df[[x for x in final_df if x not in ['readmitted']]+ ['readmitted']]
```

<a name="6"></a>
#6 - Predictive ML Models

* To predict which patients might be at high risk for readmission, we tested three predictive models: Decision Tree, Random Forest, and Logistic Regression

<a name="6.0"></a>
##6.0 - Train/Test Split:

* Before training the models, we divided our data into training and testing sets, reserving 20% for testing to validate our models. We also employed stratification on our target variable to ensure consistent distribution between the training and testing sets.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
#Creating the train and test sets, with a 80%/10% split
X = final_df.drop(columns='readmitted')
y = final_df['readmitted']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
```

<a name="*6.1*"></a>
## 6.1 - Random Forest

* Accuracy: 88.7%
* Recall: 1%

```python
rf = RandomForestClassifier(n_estimators=50, bootstrap=True, max_depth=None, max_features='sqrt', class_weight='balanced', random_state=1)
#Fit the model to the training data
rf.fit(X_train, y_train)

#Use the fitted model to predict y-values
y_pred_rf = rf.predict(X_test)

#Create a confusion matrix to display the tradeoff between false positives and false negatives
cm = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred_rf))
cm.plot(cmap='Greys', colorbar=False)
plt.title('Random Forest Confusion Martix')

#Also check the accuracy and recall of the model
print(f'Accuracy = {accuracy_score(y_test, y_pred_rf)}')
print(f'Recall = {recall_score(y_test, y_pred_rf)}')
#print(f'ROC_AUC = {roc_auc_score(y_test, y_pred_rf)}')
```

```python
#Checking feature importance
feature_importance = pd.Series(rf.feature_importances_, index=X_train.columns).sort_values(ascending=False)

plt.figure(figsize=(8,6))
feature_importance.head(5).plot(kind='bar')
plt.xticks(rotation=90)
plt.title('Feature Importance From Random Forest Model')
plt.ylabel('Feature Importance');
```

<a name="6.2"></a>
## 6.2 - Decision Tree
* Accuracy: 80%
* Recall: 18.1%



```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
params = {"criterion": "gini", "random_state": 1}

# Create DecisionTreeClassifier
clf = DecisionTreeClassifier(**params)

# Train classifier
clf = clf.fit(X_train, y_train)

# Predict outcome of test set
test_predictions = clf.predict(X_test)
```

```python
#Create a confusion matrix to display the tradeoff between false positives and false negatives
cm = ConfusionMatrixDisplay(confusion_matrix(y_test, test_predictions))
cm.plot(cmap='Blues', colorbar=False)
plt.title('Decision Tree Confusion Martix')

#Also check the accuracy and recall of the model
print(f'Accuracy = {accuracy_score(y_test, test_predictions)}')
print(f'Recall = {recall_score(y_test, test_predictions)}')
#print(f'ROC_AUC = {roc_auc_score(y_test, test_predictions)}')
```

```python
# Visualize decision tree
plt.figure(figsize = (16, 12))
tree.plot_tree(clf, feature_names = X.columns, max_depth = 2, fontsize = 12);
```

```python
#Checking feature importance
feature_importance = pd.Series(clf.feature_importances_, index=X_train.columns).sort_values(ascending=False)

plt.figure(figsize=(8,6))
feature_importance.head(5).plot(kind='bar')
plt.xticks(rotation=90)
plt.title('Feature Importance From Decision tree')
plt.ylabel('Feature Importance');
```

##7.1 - Challenges

Throughout our data manipulation and analysis work, we encountered the following challenges:

- While examining null values, we identified the presence of multiple representations for null values, including 'Not available', 'Missing', '?', and 'Invalid'. To maintain consistency, we opted to replace all instances of null values with 'None'.

- To ensure the proper functioning of our models, we needed to transform nominal variables into dummy variables.

- Our limited medical background resulted in some additional time spent on research to create variables that aggregated certain data points. One example of this was the grouping of diagnoses into higher-level categories.

- When dealing with diabetic medications, we observed values like 'No,' 'Steady,' 'Down,' and 'Up.' For our analysis, we chose to convert 'No' to 0 and 'Steady,' 'Down,' and 'Up' to 1.

## 7.2 - Conclusion

After exploring the demographic information of patients, we learned that the mean of inpatient stay is 5 days, with only 5.4% of them staying over 10 days. Among these records, the ones that left between 8-10 days see the highest re-admitted rate at about 14%, higher than the mean of 12%.

Among these records, we also found that age group [20, 30) had the highest re-admitted rate. This group of patients happen to have a shorter inpatient stay at about 3.5 days. However, the most re-admitted age group is [70, 80), as they are the largest patient population in the database.

The majority of the records are patients from Caucasian race. Although there is no significant difference in re-admission rate among ethnicity groups, Asians and Hispanics see a slightly lower re-admitted rate at around 10%.

-------

We employed three different machine learning algorithms in our study: Decision Tree, Random Forest, and Logistic Regression. While each model demonstrated good accuracy levels on our test set, a challenge surfaced. Their accuracy (~80%), was counterbalanced by a significant False Negative rate. Essentially, the models predominantly predicted a '0' (i.e., the patient will not be readmitted). This led to a deficiency in fulfilling the primary objective of our analysis, which is to accurately pinpoint patients at a higher risk of readmission.

For a more nuanced understanding, we measured the Recall for each model, which reflects their capacity to correctly identify patients likely to be readmitted '1'. The Decision Tree emerged as the most adept, with a recall of 18%.In our exploration of predictor significance, we visualized the top splits of the Decision Tree in Section 7.2, while for the Random Forest, we rendered a Feature Importance graph in Section 7.1. For the Random Forest, the top three predictors were the same, indicating that the number of lab tests, medications and lenght of stay might be important variables when trying to classify patients as high or low risk of readmission.

In conclusion, while the current models' recall rates did not meet our aspirational standards, they did shine a light on important variables that are likely important in predicting patient readmissions. Looking ahead, we identified a few areas for improvement:

Hyperparameter optimization.
Exploring alternative variable transformations.
Addressing patient record redundancies, either by aggregation or retaining a singular record, to ensure data independence.
